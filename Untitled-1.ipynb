```python
# Constants
V = 0.002  # m^3

p1_min = data['p_1'].min()

# Calculating S(t) for the entire dataset
data['S'] = np.where(data['t'] != 0, V / data['t'] * np.log(data['p_1'] / p1_min), np.nan)

# Splitting the data into different segments based on point numbers
data_0_to_50 = data.iloc[0:51]
data_51_to_100 = data.iloc[51:101]
data_101_to_end = data.iloc[101:]

# Calculating S(t) for the segmented data
data_0_to_50['S'] = np.where(data_0_to_50['t'] != 0, V / data_0_to_50['t'] * np.log(data_0_to_50['p_1'] / p1_min), np.nan)
data_51_to_100['S'] = np.where(data_51_to_100['t'] != 0, V / data_51_to_100['t'] * np.log(data_51_to_100['p_1'] / p1_min), np.nan)
data_101_to_end['S'] = np.where(data_101_to_end['t'] != 0, V / data_101_to_end['t'] * np.log(data_101_to_end['p_1'] / p1_min), np.nan)

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot for full data
axs[0, 0].plot(data['t'], data['S'], label='S(t)',  linewidth=1,marker='o')
axs[0, 0].set_title('0 to end')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot for data from 0 to 50
axs[0, 1].plot(data_0_to_50['t'], data_0_to_50['S'], label='S(t)',linewidth=1,marker='o')
axs[0, 1].set_title('0 to 50')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot for data from 51 to 100
axs[1, 0].plot(data_51_to_100['t'], data_51_to_100['S'], label='S(t)',linewidth=1,marker='o')
axs[1, 0].set_title('51 to 100')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot for data from 101 to end
axs[1, 1].plot(data_101_to_end['t'], data_101_to_end['S'], label='S(t)',linewidth=1,marker='o')
axs[1, 1].set_title('101 to end')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Setting labels for all axes
for ax in axs.flat:
    ax.set(xlabel='Время, c', ylabel='S(t), м^3/c')

# Layout adjustment
plt.tight_layout()

# Save the plot to a file
output_file_path = 'garf_2.png'
plt.savefig(output_file_path)

# Display the plot
plt.show()


```

```python
from scipy.stats import linregress
# Perform linear regression to find the slope (angle of inclination) for p_2(t)
slope, intercept, r_value, p_value, std_err = linregress(data['t'], data['p_2'])

# The slope of the line is the tangent of the angle of inclination
angle_of_inclination = np.arctan(slope)

# Convert the angle from radians to degrees
angle_of_inclination_degrees = np.degrees(angle_of_inclination)
slope, angle_of_inclination_degrees
```
